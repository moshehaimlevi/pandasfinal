```python
import numpy as np
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/mpg.csv'
df = pd.read_csv(file_path)
display(df)
```

```python
########################### SEABORN SECTION ################################


########################### QUESTION NUMBER 1 ################################
plt.figure(figsize=(15,5))

sns.scatterplot(data=df, x='horsepower', y='mpg', hue='origin', palette='Spectral', alpha=0.9, marker='o')

plt.tight_layout()
plt.legend(loc=(1.1, 0))

plt.show()
```

```python
########################### QUESTION NUMBER 2 ################################
cylinder_df = df[df['cylinders'].isin([4, 8])]

plt.figure(figsize=(8, 5))
sns.barplot(data=cylinder_df, x='cylinders', y='mpg', hue='origin', palette='Spectral', alpha=0.9)
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.title('MPG: 4 vs 8 Cylinders')
plt.tight_layout()
plt.show()
```

```python
########################### QUESTION NUMBER 3 ################################

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='origin', hue='origin', palette='Spectral')
plt.title('Vehicle Count by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```

```python
########################### QUESTION NUMBER 4 ################################
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='weight', bins=30, kde=True, color='green')
plt.title('production of vehicle weights')
plt.xlabel('weight')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
########################### QUESTION NUMBER 5 ################################
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='cylinders', y='mpg', palette='Spectral')
plt.title('MPG Distribution by Number of Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.tight_layout()
plt.legend()
plt.show()
```
