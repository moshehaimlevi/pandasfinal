```python

df_books['non_nan_count'] = df_books.notna().sum(axis=1)

df_books.sort_values(['catalog_number', 'non_nan_count'], ascending=[True, False], inplace=True)

duplicates = df_books[df_books['catalog_number'].duplicated(keep=False)]
display(duplicates)
```

```python
kept = duplicates.drop_duplicates(subset=['catalog_number'], keep='first')
removed = duplicates[~duplicates['catalog_number'].isin(kept['catalog_number'])]
display(removed)
display(kept)

duplicates.drop_duplicates(subset=['catalog_number'], keep='first', inplace=True)
df_books_remove = df_books[~df_books['catalog_number'].isin(removed['catalog_number'])]
display(df_books_remove)
```
