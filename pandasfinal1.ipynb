```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
############################ SECTION A ############################

############################ LINE CHART ############################
import matplotlib.pyplot as plt
years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

plt.plot(years, height, label='height')

plt.title(" Years V heights ")
plt.xlabel('Years')
plt.ylabel('Height')

plt.legend()
plt.xlim(2015, 2019)
plt.ylim(110, 133)
plt.xticks(years)
plt.show()
```

```python
############################ SUBPLOT ############################
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(months, temperature, color='orange', marker='o')
plt.title("Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temp (Â°C)")

plt.subplot(1, 2, 2)
plt.bar(months, rain, color='blue')
plt.title("Monthly Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.tight_layout()

plt.show()

```

```python
############################ HISTOGRAM ############################
import matplotlib.pyplot as plt

grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]

plt.figure(figsize=(5, 5))

plt.hist(grades, bins=10, color='green', edgecolor='black')

plt.title("Distribution of Grades")
plt.xlabel("Grade")
plt.ylabel("Number of Students")

plt.show()
```

```python
############################ BAR CHART ###########################
import matplotlib.pyplot as plt

age_groups = ['10-12', '13-15', '16-18', '19-21']
students_count = [30, 45, 25, 10]

plt.figure(figsize=(10, 5))

plt.bar(age_groups, students_count, color='green', edgecolor='black')

plt.title("Number of Students by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Students")

plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```
