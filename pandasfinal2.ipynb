```python
import numpy as np
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/studnet_course.csv'
df_studnet_course = pd.read_csv(file_path)
display(df_studnet_course)
```

```python
df_miss = df_studnet_course.dropna()
display(df_miss)
```

```python
df_drop = df_studnet_course.drop_duplicates()
display(df_drop)
```

```python
########################### QUESTION NUMBER 1 ################################
gender_counts = df_students['gender'].value_counts()
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', color='yellow', edgecolor='black')

plt.title('Number of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='-', alpha=0.9)

plt.show()
```

```python
########################### QUESTION NUMBER 2 ################################
plt.figure(figsize=(8, 5))
plt.hist(df_students['age'], bins=50, color='blue', edgecolor='white')
plt.title("Age Distribution of Students")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle='-', alpha=0.9)
plt.show()
```

```python
########################### QUESTION NUMBER 3 ################################
course_count = df_courses['course_category'].value_counts()
plt.figure(figsize=(10, 6))
course_count.plot(kind='bar', color='blue', edgecolor='white')

plt.title('course categories')
plt.xlabel('categories')
plt.ylabel('Number of courses in the category')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='-', alpha=0.9)

plt.show()
```

```python
########################### QUESTION NUMBER 4 ################################
city_counts_max = df_students['city'].value_counts()
most_students_count = city_counts_max.max()
display(most_students_count)

most_students_city = city_counts_max.idxmax()
display(most_students_city)
```

```python
city_counts_min = df_students['city'].value_counts()
least_students_count = city_counts_min.min()
display(least_students_count)

least_students_city = city_counts_min.idxmin()
display(least_students_city)
```

```python
########################### QUESTION NUMBER 5 ################################
category_courses = df_courses.groupby('course_category')['course_name'].unique().reset_index()
display(category_courses)
```
